```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.metrics import accuracy_score, classification_report
import pandas as pd
```

```python
data = pd.read_csv('NSTU_Students_data_transformed.csv')
target = data['target']
data = data.drop(columns=['target'])
label_encoder = LabelEncoder()
target = label_encoder.fit_transform(target)

X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.2, random_state=42)

results = {}
params = {}

print("До увеличения датасета")
```

# До увеличения датасета

## LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

param_grid = {
    'penalty': ['l1', 'l2', 'elasticnet', 'none'],
    'C': [0.01, 0.1, 1, 10, 100],
    'solver': ['newton-cg', 'lbfgs', 'liblinear', 'saga'],
    'max_iter': [1000, 1100, 1500]
}

log_reg = LogisticRegression()

grid_search_log_reg = GridSearchCV(log_reg, param_grid, cv=5, scoring='accuracy')
grid_search_log_reg.fit(X_train, y_train)

print("Best parameters for Logistic Regression:", grid_search_log_reg.best_params_)
params["Best parameters for Logistic Regression:"] = grid_search_log_reg.best_params_

y_pred_log_reg = grid_search_log_reg.predict(X_test)
print("Logistic Regression Accuracy:", accuracy_score(y_test, y_pred_log_reg))
print("Classification Report for Logistic Regression:\n", classification_report(y_test, y_pred_log_reg))

results['LogReg'] = accuracy_score(y_test, y_pred_log_reg)
```

## Random forest

```python
from sklearn.ensemble import RandomForestClassifier

param_grid_rf = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False]
}

rf_clf = RandomForestClassifier(random_state=42)

grid_search_rf = GridSearchCV(rf_clf, param_grid_rf, cv=5, scoring='accuracy')
grid_search_rf.fit(X_train, y_train)

params["Best parameters for Random Forest:"]= grid_search_rf.best_params_

y_pred_rf = grid_search_rf.predict(X_test)
print("Random Forest Accuracy:", accuracy_score(y_test, y_pred_rf))
print("Classification Report for Random Forest:\n", classification_report(y_test, y_pred_rf))

results['RandForest'] = accuracy_score(y_test, y_pred_rf)

```

## XGBoost

```python
import xgboost as xgb

param_grid_xgb = {
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 5, 7],
    'n_estimators': [100, 200, 300, 500],
    'subsample': [0.8, 1.0]
}

xgb_clf = xgb.XGBClassifier(use_label_encoder=False, eval_metric='mlogloss')

grid_search_xgb = GridSearchCV(xgb_clf, param_grid_xgb, cv=5, scoring='accuracy')
grid_search_xgb.fit(X_train, y_train)

print("Best parameters for XGBoost:", grid_search_xgb.best_params_)
params["Best parameters for XGBoost:"] = grid_search_xgb.best_params_

y_pred_xgb = grid_search_xgb.predict(X_test)
print("XGBoost Accuracy:", accuracy_score(y_test, y_pred_xgb))
print("Classification Report for XGBoost:\n", classification_report(y_test, y_pred_xgb))

results['XGB'] = accuracy_score(y_test, y_pred_xgb)
```

## KNeighborsClassifier

```python
from sklearn.neighbors import KNeighborsClassifier

param_grid_knn = {
    'n_neighbors': [3, 5, 7, 9],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan', 'minkowski']
}

knn_clf = KNeighborsClassifier()

grid_search_knn = GridSearchCV(knn_clf, param_grid_knn, cv=5, scoring='accuracy')
grid_search_knn.fit(X_train, y_train)

print("Best parameters for KNN:", grid_search_knn.best_params_)
params["Best parameters for KNN:"] = grid_search_knn.best_params_

y_pred_knn = grid_search_knn.predict(X_test)
print("KNN Accuracy:", accuracy_score(y_test, y_pred_knn))
print("Classification Report for KNN:\n", classification_report(y_test, y_pred_knn))

results['KNN'] = accuracy_score(y_test, y_pred_knn)
```

```python
from scipy.stats import pearsonr
def strong_significant_correlation(df, threshold=0.7, p_value_threshold=0.05):
    """
    Функция для проверки на сильную и статистически значимую корреляцию между атрибутами.
    
    :param df: DataFrame с данными
    :param threshold: Порог для сильной корреляции (по умолчанию 0.7)
    :param p_value_threshold: Порог p-значения для статистической значимости (по умолчанию 0.05)
    :return: DataFrame с парами атрибутов, которые имеют сильную и значимую корреляцию
    """
    df_copy = df.copy()
    
    correlations = []

    df_copy = df_copy.astype(float)

    for i in range(len(df_copy.columns)):
        for j in range(i + 1, len(df_copy.columns)):
            feature1 = df_copy.columns[i]
            feature2 = df_copy.columns[j]

            corr, p_value = pearsonr(df_copy[feature1], df_copy[feature2])

            if abs(corr) > threshold and p_value < p_value_threshold:
                correlations.append((feature1, feature2, corr, p_value))

    result_df = pd.DataFrame(correlations, columns=['Attribute 1', 'Attribute 2', 'Correlation Coefficient', 'P-Value'])
    return result_df
```
