```python
import pandas as pd
```

```python
data = pd.read_csv('NSTU_Students_data_transformed.csv')
data.head(5)
```

```python
data.corr
```

```python
target = data['target']
data = data.drop(columns=['target'])
```

```python
data.head(4)
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
target = label_encoder.fit_transform(target)
```

```python
target
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(X_train, y_train)
```

```python
X_test
```

```python
y_pred = knn.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test, y_pred)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,  y_pred)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_searcher = GridSearchCV(KNeighborsClassifier(),
                             param_grid={'n_neighbors': range(1, 100, 2),
                                         'weights': ['uniform', 'distance'],
                                         'p': [1, 2, 3]},
                             cv=5)
```

```python
grid_searcher.fit(X_train, y_train)
```

```python
best_predictions = grid_searcher.predict(X_test)
```

```python
mean_squared_error(y_test, best_predictions)
```

```python
accuracy_score(y_test,  best_predictions)
```
