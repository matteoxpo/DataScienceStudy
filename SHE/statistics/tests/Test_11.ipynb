# Меры разброса

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats
import statsmodels.api as sm
import numpy as np

candy = pd.read_csv('CandyData.csv')
candy.head()
```

– chocolate/fruity/caramel/peanutalmondy/nougat/crispedricewafer — содержит ли конфета шоколад/фруктовый ароматизатор/карамель/арахис/нугу/воздушный рис (1 – да/ 0 – нет)<br>
– hard — конфета твердая (1 – да/ 0 – нет)<br>
– bar — конфета похожа на батончик (1 – да/ 0 – нет)<br>
– pluribus — в упаковке одна конфета или несколько (1 – несколько/ 0 – одна)<br>
– sugarpercent — процент сахара в конфете<br>
– pricepercent — цена конфеты<br>
– winpercent — рейтинг конфеты<br>

```python
trio = candy[['sugarpercent', 'pricepercent', 'winpercent']]
```

```python
trio.corr()
```

```python
X = candy['sugarpercent']
Y = candy['winpercent']
X_mult_constant = sm.add_constant(X)
mod = sm.OLS(Y, X_mult_constant)
res_mult = mod.fit()
print(res_mult.summary())
```

```python
p_value = res_mult.pvalues['sugarpercent']
print(f'P-value for sugarpercent: {p_value:.4f}')

alpha = 0.05

if p_value < alpha:
    print("На уровне значимости 0.05 процент сахара в конфете статистически значимо влияет на рейтинг конфеты.")
else:
    print("На уровне значимости 0.05 процент сахара в конфете не статистически значимо влияет на рейтинг конфеты.")
```

```python
X = candy[['sugarpercent', 'pricepercent']]
Y = candy['winpercent']

X = sm.add_constant(X)
model = sm.OLS(Y, X).fit()

coefficients = model.params

print(coefficients)

pricepercent_coef = coefficients['pricepercent']
sugarpercent_coef = coefficients['sugarpercent']
print(f'Коэффициент при цене конфеты: {pricepercent_coef:.2f}')
print(f'Коэффициент при проценте сахара: {sugarpercent_coef:.2f}')
```

```python
p_value = model.pvalues['sugarpercent']
print(f'P-value for sugarpercent: {p_value:.4f}')

alpha = 0.05

if p_value < alpha:
    print("На уровне значимости 0.05 процент сахара в конфете статистически значимо влияет на рейтинг конфеты.")
else:
    print("На уровне значимости 0.05 процент сахара в конфете не статистически значимо влияет на рейтинг конфеты.")
```
