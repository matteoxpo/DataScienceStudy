```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


judge = pd.read_csv('TenthCircuit.csv')
judge.head(100)
```

```python
judge['Category'].unique()
```

```python

judge['votes_sum'] = judge[['Vote1', 'Vote2', 'Vote3']].sum(axis=1)
judge['vote_type'] = judge['votes_sum'].apply(lambda x: 'liberal' if x == 3  else 'conservator' )
grouped = judge.groupby(['Category', 'vote_type']).size().unstack()
grouped.plot(kind='bar', stacked=False)
plt.title('Количество заседаний по категориям и типам голосования')
plt.ylabel('Количество заседаний')
plt.xlabel('Категория дела')
plt.show()
```

```python
judge['Date'] = pd.to_datetime(judge['Date'])
judge['Month'] = judge['Date'].dt.month
judge['Year'] = judge['Date'].dt.year

def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    elif month in [9, 10, 11]:
        return 'Autumn'

judge['Season'] = judge['Month'].apply(get_season)

grouped_mean = judge.groupby('Season').size() / judge['Year'].nunique()

print(grouped_mean)

grouped_mean.plot(kind='bar')

plt.title('Среднее число заседаний за сезон')
plt.xlabel('Сезон')
plt.ylabel('Среднее число заседаний')
plt.show()

monthly_counts = judge.groupby('Month').size()
max_month = monthly_counts.idxmax()

month_name = pd.to_datetime(str(max_month), format='%m').strftime('%B')
print(f"Месяц с наибольшим количеством заседаний: {month_name}")
```

```python
monthly_mean = judge.groupby('Month').size() / judge['Year'].nunique()

print(monthly_mean)

monthly_mean.plot(kind='bar')

plt.title('Среднее число заседаний по месяцам')
plt.xlabel('Месяц')
plt.ylabel('Среднее число заседаний')
plt.xticks(ticks=range(0, 12), labels=[
    'Январь', 'Февраль', 'Март', 'Апрель', 'Май', 'Июнь',
    'Июль', 'Август', 'Сентябрь', 'Октябрь', 'Ноябрь', 'Декабрь'
], rotation=45)

plt.show()

max_month = monthly_mean.idxmax()

month_name = pd.to_datetime(str(max_month), format='%m').strftime('%B')
print(f"Месяц с наибольшим средним количеством заседаний: {month_name}")
```

```python
judge = pd.read_csv('TenthCircuit.csv')
all_judge = judge.dropna(subset=['Judge1', 'Judge2', 'Judge3'])
print(all_judge.size, judge.size)
imigr = all_judge[ all_judge['Category'] == 'Immigration']
judge_grouped = imigr.groupby('Judge1').size() + imigr.groupby('Judge2').size() + imigr.groupby('Judge3').size()

fig, ax = plt.subplots(figsize=(10,6))

ax.bar(judge_grouped.index, judge_grouped)
print(judge_grouped)
```

```python
immigration_cases = judge[judge['Category'] == 'Immigration']

all_judges = pd.concat([immigration_cases['Judge1'],
                        immigration_cases['Judge2'],
                        immigration_cases['Judge3']])

all_judges = all_judges.dropna()

judge_grouped = all_judges.value_counts()

fig, ax = plt.subplots(figsize=(10,6))
judge_grouped.plot(kind='bar', ax=ax)

plt.title('Количество заседаний по каждому судье (Immigration)')
plt.xlabel('Судья')
plt.ylabel('Количество заседаний')
plt.show()

top_judge = judge_grouped.idxmax()
print(f"Судья, который принял участие в наибольшем количестве заседаний: {top_judge}")
```

```python
mode_title = judge['Title'].mode()
mode_title_j = judge[judge['Title'].isin(mode_title)].sort_values(by='Title')

all_judges = pd.concat([mode_title_j['Judge1'],
                        mode_title_j['Judge2'],
                        mode_title_j['Judge3']])
all_judges = all_judges.dropna()
judge_grouped = all_judges.value_counts()
fig, ax = plt.subplots(figsize=(10,6))
judge_grouped.plot(kind='bar', ax=ax)

plt.title('Количество заседаний по каждому судье (Immigration)')
plt.xlabel('Судья')
plt.ylabel('Количество заседаний')
plt.show()

top_judge = judge_grouped.idxmax()
print(f"Судья, который принял участие в наибольшем количестве заседаний: {top_judge}")
```
