# Меры разброса

```python
import pandas as pd
import numpy as np


usa = pd.read_csv('UrbanizationCensusTract.csv')
usa.head()
```

```python
usa[ usa['state'] == 'Wyoming']['log_pop'].dropna().var(ddof=1)
```

```python
usa[usa['state'] == 'South Dakota']['log_pop'].dropna().var(ddof=1)
```

```python
X = usa[usa['state'] == 'South Dakota']['log_pop']
Y = usa[usa['state'] == 'Wyoming']['log_pop']
```

```python
var1 = X.var()
var2 = Y.var()
std1 = X.std()
std2 = Y.std()
print(var1, std1**2)
m1 = X.mean()
m2 = Y.mean()
v1 = var1 / len(X)
v2 = var2 / len(Y)
df = (v1 + v2) ** 2 / (  v1**2/(len(X) - 1) + v2**2/(len(Y) - 1) )
print(int(df))
```

```python
print(X.var(ddof=1) / np.sqrt(len(X)), Y.var(ddof=1) / np.sqrt(len(Y)))
```

```python
np.sqrt(var1 / len(X) + var2 /len(Y))
```

```python
Y.mean() - X.mean()
```

```python
s_m1_m2 = np.sqrt(var1/len(X) + var2/len(Y))
t = (m1 - m2)/s_m1_m2
print(t)
```

```python
import scipy.stats

scipy.stats.ttest_ind(X, Y, equal_var=False)
```

```python
len(X)+len(Y) - 2
```

```python
t_critical = 2.5920
CI_low = abs(m1 - m2) - t_critical * s_m1_m2
CI_up = abs(m1 - m2) + t_critical * s_m1_m2
print(CI_low, CI_up)
```
