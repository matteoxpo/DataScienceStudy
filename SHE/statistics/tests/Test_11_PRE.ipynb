# Меры разброса

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats
import statsmodels.api as sm
import numpy as np
```

```python
X = np.array([3, 2, 98, 34, 11, 2, 3]).reshape(-1, 1)
Y = np.array([2, 1, 0, 3, 7, 2, 15])
```

```python
X_mult_constant = sm.add_constant(X)
mod = sm.OLS(Y, X_mult_constant)
res_mult = mod.fit()
res_mult.summary()
```

```python
from sklearn.metrics import mean_squared_error
Y_pred = res_mult.predict(X_mult_constant)
mse = mean_squared_error(Y, Y_pred)
mse
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(Y, Y_pred)
mae
```

```python
plt.figure(figsize=(20, 6))
plt.scatter(X, Y, color='blue', label='Исходные данные')
plt.plot(X, Y_pred, color='red', label='Линия регрессии')

plt.title('Линейная регрессия')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()

plt.show()
```
