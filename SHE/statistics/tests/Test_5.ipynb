# Меры разброса

```python
import pandas as pd
import numpy as np

marvel = pd.read_csv('MarvelComicCaracters.csv', sep=';')
marvel.head()
```

```python
cleared_marvel = marvel.dropna(subset='appearances')
cleared_marvel
```

```python
cleared_marvel[ cleared_marvel['align'] == 'Злой персонаж'].describe()
```

```python
q1 = cleared_marvel['appearances'].quantile(0.25)
q3 = cleared_marvel['appearances'].quantile(0.75)
iqr = q3 - q1
iqr
```

```python
iqr_outlier_threshold_bottom = q1 - 1.5 * iqr
iqr_outlier_threshold_up = q3 + 1.5 * iqr
print(iqr_outlier_threshold_bottom)
print(iqr_outlier_threshold_up)
```

```python
outsiders = cleared_marvel[cleared_marvel['appearances'] > iqr_outlier_threshold_up]
print(outsiders.shape)
# print(outsiders.shape)
print(outsiders[outsiders['align'] == 'Добрый персонаж'].shape)
print(outsiders[outsiders['align'] == 'Нейтральный персонаж'].shape)
print(outsiders[outsiders['align'] == 'Злой персонаж'].shape)
```

```python
cleared_marvel = marvel.dropna(subset='appearances')
clear_ln = np.log(cleared_marvel['appearances'])
std = clear_ln.std()
mean = clear_ln.mean()
print(std, mean, sep=' ')
```

```python
std_outlier_threshold_bottom = mean - 3 * std
std_outlier_threshold_up = mean + 3 * std
print(std_outlier_threshold_bottom)
print(std_outlier_threshold_up)
```

```python
filtered_clear_ln = clear_ln[(clear_ln > std_outlier_threshold_bottom) & (clear_ln < std_outlier_threshold_up)]

# Подсчитываем количество выбросов
num_outliers = clear_ln.shape[0] - filtered_clear_ln.shape[0]
num_outliers
```

```python
cleared_marvel = marvel.dropna(subset='appearances')
std = cleared_marvel['appearances'].std()
mean = cleared_marvel['appearances'].mean()
std_outlier_threshold_bottom = mean - 3 * std
std_outlier_threshold_up = mean + 3 * std
print(std_outlier_threshold_bottom)
print(std_outlier_threshold_up)
```

```python
filtered_cleared = cleared_marvel[(cleared_marvel['appearances'] > std_outlier_threshold_bottom) & (cleared_marvel['appearances'] < std_outlier_threshold_up)]
cleared_marvel.shape[0] - filtered_cleared.shape[0]
```

```python
cleared_marvel = marvel.dropna(subset='appearances')
cleared_marvel = cleared_marvel[cleared_marvel['sex'] == 'Женский персонаж']['appearances']
cleared_marvel
```

```python
M = cleared_marvel.mean()
sigma = cleared_marvel.std()
def z_score(X, M, sigma):
  return (X - M) / sigma
z_scores = cleared_marvel.apply(lambda x: z_score(x, M, sigma))
z_scores
```

```python
mean = z_scores.mean()
print(mean)
std = z_scores.std()
print(std)
std_outlier_threshold_bottom = mean - 3 * std
std_outlier_threshold_up = mean + 3 * std
print(std_outlier_threshold_bottom)
print(std_outlier_threshold_up)
```

```python
filtered_cleared = z_scores[(z_scores > std_outlier_threshold_bottom) & (z_scores < std_outlier_threshold_up)]
z_scores.shape[0] - filtered_cleared.shape[0]
```
