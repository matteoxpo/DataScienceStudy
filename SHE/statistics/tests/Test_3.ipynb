```python
import pandas as pd

marvel = pd.read_csv('MarvelComicCaracters.csv', sep=';')
print(marvel.shape)
marvel.head()
```

```python
appear = marvel['first_appearance'].unique()
count = -1
year = appear[0]
for app in appear:
  curr_count = marvel[marvel['first_appearance'] == app].shape[0]
  if curr_count > count:
    count = curr_count
    year = app
print(year, count, sep= " ")
```
