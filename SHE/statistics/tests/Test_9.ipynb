# Меры разброса

```python
import pandas as pd
import numpy as np


star = pd.read_csv('StarWars.csv', sep=';')
star.head()
```

```python
ct_o = star.groupby('gender')['fan'].value_counts().unstack()
ct_o
```

```python
n_total = ct_o.sum().sum()
ct_o['да'].sum() * ct_o.iloc[0].sum() / n_total
```

```python
ct_o['нет'].sum() * ct_o.iloc[1].sum() / n_total
```

```python
import scipy.stats
```

```python
print(scipy.stats.chi2_contingency(ct_o, correction=False))
```

```python
cleared = star.dropna(subset='fan')
len(cleared)
len(cleared[cleared['fan'] == 'да']) / len(cleared)

# len(star[star['fan'] == 'да']) / len(star)
```
