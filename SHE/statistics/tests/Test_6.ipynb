```python
df_tie_1 = games[games['tie'] == 1]
df_tie_0 = games[games['tie'] == 0]
```

```python
grouped_tie_1 = df_tie_1.groupby('tourneyid').agg({
    'winnernewrating': 'mean',
    'losernewrating': 'mean'
})

# Группировка по tourneyid и вычисление средних значений для tie = 0
grouped_tie_0 = df_tie_0.groupby('tourneyid').agg({
    'winnernewrating': 'mean',
    'losernewrating': 'mean'
})

# Рассчитайте коэффициент корреляции для игр с ничьей (tie = 1)
correlation_tie_1 = grouped_tie_1[['winnernewrating', 'losernewrating']].corr().loc['winnernewrating', 'losernewrating']

# Рассчитайте коэффициент корреляции для игр без ничьей (tie = 0)
correlation_tie_0 = grouped_tie_0[['winnernewrating', 'losernewrating']].corr().loc['winnernewrating', 'losernewrating']
print(correlation_tie_1)
print(correlation_tie_0)
```
