```python
df['sugarpercent_centered'] = df['sugarpercent'] - df['sugarpercent'].mean()
df['pricepercent_centered'] = df['pricepercent'] - df['pricepercent'].mean()

X = df[['sugarpercent_centered', 'pricepercent_centered']]
y = df['winpercent_0_1']

X_constant = sm.add_constant(X)

logit = sm.Logit(y, X_constant).fit()
print(logit.summary())
```

```python
sugar_mean_centered = df['sugarpercent'].mean()
price_mean_centered = df['pricepercent'].mean()

coef_const = logit.params['const']
coef_sugar = logit.params['sugarpercent']
coef_price = logit.params['pricepercent']

logit_value = coef_const + coef_sugar * sugar_mean_centered + coef_price * price_mean_centered

odds = np.exp(logit_value)
round(odds,2)
```

```python
df = pd.read_csv('CandyData.csv')
df['winpercent_0_1'] = df['winpercent'].apply(lambda x: 1 if x >= 50 else 0)

df['sugarpercent_centered'] = df['sugarpercent'] - df['sugarpercent'].mean()
df['pricepercent_centered'] = df['pricepercent'] - df['pricepercent'].mean()

X = df[['sugarpercent', 'pricepercent']]
X_constant = sm.add_constant(X)
y = df['winpercent_0_1']

logit = sm.Logit(y, X_constant).fit()

print(logit.summary())

coef_const = logit.params['const']
coef_sugar = logit.params['sugarpercent']
coef_price = logit.params['pricepercent']

sugar_mean_centered = df['sugarpercent'].mean()
price_mean_centered = df['pricepercent'].mean()

logit_value = coef_const + coef_sugar * sugar_mean_centered + coef_price * price_mean_centered

odds = np.exp(logit_value)

print(f"Odds at average sugarpercent and average pricepercent: {odds}")
```
