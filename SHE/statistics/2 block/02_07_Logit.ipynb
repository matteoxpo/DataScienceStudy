```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
import numpy as np

df = pd.read_csv('heart.csv')
df.head()
```

```python
cols_num = ['age', 'trestbps', 'chol', 'thalach']
```

```python
X = df['sex']
y = df['target']
X_constant = sm.add_constant(X)
logit = sm.Logit(y, X_constant).fit()
print(logit.summary())
```

```python
t_s = df.groupby('target')['sex'].value_counts().unstack()
t_s
```

```python
p_1_male = t_s[1][1] / (t_s[1][1] + t_s[1][0])
p_0_male = 1 - p_1_male
odds_1_male = p_1_male / p_0_male
print(odds_1_male)
print(np.log(odds_1_male))
```

```python
p_1_female = 72 / (72 + 24)
p_0_female = 1 - p_1_female
odds_1_female = p_1_female / p_0_female
print(odds_1_female)
print(np.log(odds_1_female))
```

```python
odds_fem_male = (72 / 24) / (93 / 114)
print(odds_fem_male)
print(np.log(odds_fem_male))
```

```python
X = df.drop('target', axis=1)
y = df['target']
X_constant = sm.add_constant(X)
logit = sm.Logit(y, X_constant).fit()
print(logit.summary())
```

```python
X = df[cols_num]
y = df['target']
X_constant = sm.add_constant(X)
logit = sm.Logit(y, X_constant).fit()
print(logit.summary())
```

```python
X = df[cols_num] - df[cols_num].mean()
y = df['target']
X_constant = sm.add_constant(X)
logit = sm.Logit(y, X_constant).fit()
print(logit.summary())
```

```python
X = (df[cols_num] - df[cols_num].mean()) / df[cols_num].std()
y = df['target']
X_constant = sm.add_constant(X)
logit = sm.Logit(y, X_constant).fit()
print(logit.summary())
```
