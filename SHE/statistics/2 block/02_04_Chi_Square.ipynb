# Критерий хи-квадрта

```python
import pandas as pd
import numpy as np
import scipy.stats

df = pd.read_csv('heart.csv')
df.head()
```

```python
ct_o = df.groupby('sex')['target'].value_counts().unstack()
ct_o
```

```python
df.groupby('sex')['target'].value_counts()
```

Смотрим на пол и на таргет
нулевая гепотеза: наблюдения распределны по категориям и они независимы

$
dof = (R - 1)(C - 1)
$

```python
dof = (ct_o.shape[0] - 1) * (ct_o.shape[1] - 1)
dof
```

```python
alpha = 0.01
critical_value = 6.63
```

```python
n_total = ct_o.sum().sum()
print(n_total)
```

```python
n_target_0 = ct_o[0].sum()
n_target_1 = ct_o[1].sum()
print(n_target_0, n_target_1)
```

```python
p_target_0 = n_target_0 / n_total
p_target_1 = n_target_1 / n_total
print(p_target_0, p_target_1)
```

```python
n_women = ct_o.iloc[0].sum()
n_men = ct_o.iloc[1].sum()
print(n_women, n_men)
```

```python
f_e_target_0_women = n_women * p_target_0
f_e_target_1_women = n_women * p_target_1
print(f_e_target_0_women, f_e_target_1_women)
```

```python
f_e_target_0_men = n_men * p_target_0
f_e_target_1_men = n_men * p_target_1
print(f_e_target_0_men, f_e_target_1_men)
```

```python
print(f_e_target_0_women + f_e_target_1_women + f_e_target_0_men + f_e_target_1_men)
```

```python
print((f_e_target_0_women + f_e_target_1_women + f_e_target_0_men + f_e_target_1_men) == n_total)
```

```python
ct_e = np.array([[f_e_target_0_women, f_e_target_1_women], [f_e_target_0_men, f_e_target_1_men]])
```

```python
print(ct_e)
```

$
f_e = \frac{f_cf_r}{n}
$

```python
ct_o
```

```python
ct_o[0].sum() * ct_o.iloc[0].sum() / n_total
```

```python
ct_o.iloc[0].sum()
```

```python
ct_o[0].sum()
```

$
\chi^2 = \sum{\frac{(f_o - f_e)^2}{f_e}}
$

```python
chi_square = ((ct_o - ct_e)**2/ct_e).sum().sum()
print(chi_square)
```

```python
print(chi_square > critical_value)
```

```python
print(scipy.stats.chi2_contingency(ct_o, correction=False))
```
