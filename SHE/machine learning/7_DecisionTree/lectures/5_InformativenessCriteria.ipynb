  
Использовали энтропию для оценки распределений  
По сути это мера "хаотичности" "засоренности" - Impurity  

Рассмотрим другой пример  
  
Интерпретация:
1. Пусть мы взяли какую-то определенную вершину дерева
   Случайный классификатор выдает значение объекту, что он относится к классу $k$ с вероятностью $p_k$  
   То веротяность ошибки такого классификатора будет равна критерию Джини  
   Если в вершине находится всего 1 класс, то логично ошибок не будет  
   Если распределение будет равномерное, то ошибка будет максимальная
2. Если вместо $p_i$ стояло бы количество объектов i-го класса, то формула равна количеству пар таких объектов из вершины, что эти объекты приходят из разных классов  
   Получается доля таких пар-объектов, что эти объекты из разных классов  
   Но по факту у нас вероятности...  
   *Я не понял вообще эту интерпретацию*


Пусть у нас всего 2 класса  
  
Видно, что энтропия и Джини очень похожи, минимум значения при низком количестве объектов одного класса, максимум при равномерном распределении  
Очень редко один эффективнее другого по значимости, но критерий Джини по алгоритмам быстрее  
  
  
  
$R$ - текущая вершина $R_l$ - левое $R_r$ - правое  
$j$ - какой признак  $t$ - какой порог
