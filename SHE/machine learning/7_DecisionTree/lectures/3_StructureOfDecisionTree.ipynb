$R_v$ - объекты в области $v$ 
Например для задачи регрессии логичнее всего выдавать среднее по объектам в этой области  
Для задачи классификации берем самый частый в этой области  
  
Доля объектов среди классов  
Можно делать в листьях что-угодно, например в листе может быть другая модель  
  
Аналитическая формула для дерева  
  
Каждая листовая вершина соответствует какой-то области  
  
Надо заметить, что $R_1,...,R_j$ - это листовые вершины
Договоримся, что $c_j$ - прогноз для $R_j$
$[x \in R_j]$ - индикатор, что объект попадает в область

Опа а модель то линейная  
  
По сути это скалярное произведение 
$[x \in R_j]$ - некоторые признаки, то есть этот индикатор можно свести к проверке исходных признаков на факт попадания в область $R_v$  
$c_j$ - вес в линейной модели  
Но обучать мы их будем вовсе не как линейную модель
