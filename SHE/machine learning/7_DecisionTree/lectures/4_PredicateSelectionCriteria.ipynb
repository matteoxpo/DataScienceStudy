Энтропия - мера неопределенности исхода эксперимента  
Слева на графике энтропия говорит о том, что вероятности всех событий одинаковые, после окончания эксперемента приходит много информации, а до эксперемента мы мало что знаем  
График справа - еще до начала эксперемента мы знаем, что 4 событие просиходит сильно чаще. Мы можем догадываться о том, что произойдет и скорее всего после эксперемента новой информации мы не получим  
  
Много неорпеделенности vs мало неопределенности
