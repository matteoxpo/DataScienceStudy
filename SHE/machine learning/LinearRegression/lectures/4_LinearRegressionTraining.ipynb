Обсудим как обучать линейные модели - обучение сводится к поиску вектора столбца w такого, что метрика ошибки будет минимальной

Прежде чем обсуждать как находить w вспоминим:<br>
<br>
Вспомним про свойства производный и угол касательной и тд и тп<br>
перейдем к тому, что мы будем использовать<br>
<br>
<br>
Перейдем к ГРАДИЕНТУ - когда функция зависит от многих переменных<br>
<br>
Градиент или же по-умному НАБЛА(так называется треугольник перевернутый) является вектором частных производных функции<br>
У него много свойст, но нас интересует одно<br>
Рассмотрим пример, где функция двух переменных, то есть в 3-мерном пространстве<br>
<br>
Допустим мы взяли точку $x_0$ и мы задаемся вопросом в какую сторону функция будет расти быстрее всего<br>
На нижней плоскости нарисованы градиенты точек, куда они показывают туда функция и возрастает<br>
Зафиксируем это свойство<br>
<br>
Перейдем к экстремуму функции<br>
<br>
Посчитаем градиент для нашего MSE, чтобы найти минимум от MSE<br>
<br>
в решении видно<br>
$
w = (X^T \times X)^(-1) \times y
$
Перемножение $(X^T \times X)$ дает нам квадртную матрицу, решая проблему с тем, что у нас разное количество наблюдений и признаков<br>
Дай бог эта матрица не будет вырожденной, то есть $det(X^T \times X) =0$ это плохо, либо дай бог определитель не близок к нулю<br>
<br>

Это подход **только для MSE**<br>
