Допустим мы хотим предсказывать стоимость квартиры  
  
Как опредилить какой признак важнее?  
**Чем больше вес тем важнее признак?**  
Нет, можно очень просто объяснить: допустим мы поменяем единицы измерения площади с метров на сантиметры, то кф придется уменьшить в 10 раз  
**Надо масштабировать признаки**  
  
  
После проделанных операций признаки будут в одном масштабе и можно сравнивать между собой  
<br>
$(x_i)^j$ - конкретное значение $j-го$ признака по $i-му$ наблюдению  
$
\mu - среднее
\\
\sigma - стандартное отклонение  
$  
Есть другие способы масштабирования, это один из них  
  
  
  
Получается, что второй признак важнее, так? раз перед ним 2  
  
Важно смотреть как портится модель при удалении какого-либо признака, то есть важнее все-таки первый признак, так как MSE при удалении второго признака меньше, чем при удалении первого признака
