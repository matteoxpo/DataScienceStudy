## Расстояние до точки
  
Расстояние от разных точек до $a_0$ - усреденное расстояние от вектора до точки
## Расстояние от точек $x_i$ до многообразия $L_1$
  
$a_0$ - смещение
$a_1$ - нормированный вектор, вдоль которого можно проводить линию  
Мы хотим найти длину от нашего $x_i$ до нашей прямой  
*Уберем смещение - по сути возьмем разницу векторов $x_i - a_0$  
  
Спроецируем нашу разница на вектор $a_1$  
  
На всякий случай отмечу - $(a,b)$ - скалярное произведение векторов $a$ и $b$  
$(x_i - a_0, a_1)$ - скалярное произведение вектора $x_i - a_0$ и $a_1$  
**Нас интересует** - перпендикуляр, который нарисован серым пунктиром  
то есть нас интересует разница между сними вектором и зеленым:
$$
(x_i - a_0) - ( a_1 \times (x_i - a_0, a_1) )
$$
Ну а если точнее то норма этого дела  
  
## Расстояние от $x_i$ до плооскости
  
Здесь по сути будет тоже самое:  
Находим разницу $x_i - a_0$  
Берем проекции этого вектора на $a_1$, $a_2$  
  
Заметим что проекция синего вектора $(x_i - a_0)$ на плоскость это и есть сумма проекции этого же вектора на вектора $a_1$, $a_2$:  
  
То есть нам надо найти снова сторону прямоугольника, где решение это разница синего и оранжевого вектора
