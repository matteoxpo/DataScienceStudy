# k-means
Кластеры - это центроиды, их есть в количестве k, а точки принадлежат к кластерам по расстояниям  
  
Как найти центроиды?  
  
Выбрали 4 случайных центроиды и распределим объекты по кластерам относительно ближайших объектов  
  
Наугад взять центроиды не получилось, вообще центроида должна быть более менее центровой в своем кластере, так что возьмем вариант получше  
Весь алгоритм:  
* Берем наугад k точек  
* Распределяем оставшиеся точки по расстояниям к разным центроидам  
  
* Считаем среднюю точку по всем точкам, относящимся к нашей центроиде и сдвигаем ее на это среднее
  
* Теперь связи центроид с точками невалидны, надо пересчитать расстояния и перераспределить точки по центроидам  
  
* Далее можно циклично крутить это
