в двух словах<br>
K - какая-то функция, они бывают разные, примеры ниже будут<br>
условия для функции K:<br>
K(1) дает нормальный вклад в вес $w_i$<br>
K(большое число) дает маленький вклад<br>
Так h - ширина окна, меняя которое мы можем регулировать с какими расстояниями мы будем считаться, а с какими нет<br>
h - маленькое, значит расстояние будет большое и мы мало учитываем влияние этого объекта<br>
h - большое, значит расстояние будет небольшое и мы учитываем этот объект
