Задача стоит в обучении $b_n$ так, чтобы его добавка в уже обученные модели минимизировала ошибку  
  
Проблема в том как обучать такое дело, как обучать дерево, чтобы оно было хорошей добавкой к уже существующей модели  
Попробуем рассмотреть подход  
  
Обучаем новую модель на разнице между реальным ответом и прогнозом уже построенной комопзиции  

Начнем с логистической функции потерь  
  
Логфунк:  
$L(y,z) = log(1 + exp(-yz))$  
где $y = \pm 1$, $z$ - некоторая уверенность модели в ответе  
Подставили в качестве $z b_n(x_i)$ а в качестве $y (y_i - a_{N-1}(x_i))$  
Можно заметить, что $b_n$ - должно быть числом, которое говорит об уверенности моделе в своем ответе, а знак говорить о классе, к которому относится объект  
А $(y_i - a_{N-1}(x_i))$ - классом, знак $\pm 1$  
Теперь заметим следующее:
1. Угадала модель правильно - получим $log(2)$ - то есть этот объект никак не участвует в построении новой модели  
2. Неугадала модель - получим $(y_i - a_{N-1}(x_i)) = \pm 2$, а надо $\pm 1$  

Поделим на 2: 
  
Но у такого подхода есть проблема:  
  
На обоих примерах надо, чтобы модель подкорректировала ошибки прошлого, но ошибки эти разного размаха, что не учитывается в модели, не учитывается отступ. То есть нет акцента на то, где должны быть какие отступы
