Эта лекция о способах улучшения метода бустинга  
  
Идея в том, чтобы ограничить дерево, на 1 уровень используется только 1 предикат  
Если в обычном дереве на каждой вершине перебирать всевозможные предикаты, то тут только на 1 уровне надо  
По рисунку надо всего 2 предиката перебирать  
  
Это классический способ построения дерева  
Рассмотрим другой подход  
  
Допустим есть дерево, в данном случае решающий пень  
Смотрим, разбиение какого из листьев даст наибольшее уменьшение ошибки  
Допустим это левый лист, разбили его, теперь три листа  
Смотрим из всех листьев разбиение какого уменьшит ошибку и идем так до критерия остановки (кол-во листьев)  

Посмотрим другой способ  
  
В простейшем варианте мы перебираем всевозможные значения признаков по жадному алгоритму и смотрим, какой предикат на признак самый лучший  
Это дотстаточно много, скорее всего уникальных значений признака будет столько же, сколько и объектов в обучающей выборке  
Сделаем проще, посмотрим допустим гистрограмму значений признака и будем брать парочку из столбика  

А можно еще иначе сделать  
Для каждого объекта посмотрим значение сдвига на нем, если производная по модулю большая, то это ВАЖНЫЙ объект на котором надо сильно менять прогноз; если производная около нуля, то сильно поменять прогноз не получится  

Гипотеза в варианте 3 заключается в том, что скорее всего лучшее значение порога будет среди объектов с большой по модулю производной.  
Семплируем - откидываем значения, где производная близка к нулю и ищем среди больших производных пороги  

Улучшим регуляризацию деревьев  
  
До этого мы смотрели только добавление длини шага и семлпирование объектов или признаков

За кол-во листьев  
За большие значения прогнозов  
И так далее
