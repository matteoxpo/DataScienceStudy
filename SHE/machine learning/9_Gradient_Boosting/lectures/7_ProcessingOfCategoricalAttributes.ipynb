Рассмотрим другие подходы  
  
Здесь увидим что цены в разных районах немного разные, может закодируем значение категориального признака через целевую переменную  
  
$count$ - мы считаем сколько объектов какой=то категории  
сколько раз какой-то признак встречается в выборке  
  
В задачи регрессии домножаем на число таргета, получим суммы таргетов по разным признакам  
  
В классификации посчитаем сколько встречается каждый из классов  
Еще раз, категория равна какому-то значению категории, а таргет равен тоже какому-то другому финальному значению категории  
  
Новый признак!  
Делим сумму целевых объектов с нужной категорией на количество объектов с такой категорией, по сути это *средний таргет* по всем объектам категории  
В классификации  
  
По сути это доля объектов этого класса среди всех объектов этой категории  
Посмотрим на примере  
  
ЦАО две квартиры - 10 и 9 млн, поэтому среднее между ними 9.5 и будет счетчиком
