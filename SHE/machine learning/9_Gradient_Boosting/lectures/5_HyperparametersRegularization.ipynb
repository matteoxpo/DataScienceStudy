Освежим что такое градиентный бустинг  
  
Про смещение и разборс  
  
В бэггинге:  
разброс уменьшается и зависит от того насколько модели друг от друга независимы
смещение не изменяется  
В градиентном бустинге:  
Смещение уменьшается в любом случае, даже если модели были очень слабые, то итоговая композиция будет мощная  
Разброс может увеличиться, бустинг вполне может переобучиться  
Резюмируя: градиентный бустинг лучше выбирать модели, которые меньше переобучаются, смещенные и с низким разбросом  
  
Понятно, что у деревьев есть и другие гиперпараметры, например какая функция потерь и минимальное кол-во объектов в листе, но опустим это  
Есть и другие проблемы  
  
Сдвиги $s_i$ показывают куда мы хотим сдвигать композицию, но так как базовые модели очень простые, то есть шанс того, что аппроксимация работает не очень. То есть можем даже увеличить ошибку  
Решение  
  
То есть мы теперь не так сильно доверяем нашим базовым моделям, добавляем их с небольшим весом  
  
$\eta$ - по сути скорость обучения. Мы меняем вклад новой модели.  
Чем меньше $\eta$ тем больше итераций надо  
  
Как видно по графику:  
Оранжевый - по сути мы вернулись к исходному, но мы быстро дошли до минимума ошибки и далее пошло переобучение  
При уменьшении $\eta$ надо больше итераций для поиска минимума, но и больше итераций для переобучения  
Чем меньше длина шага, тем ниже ошибка, это видно по графику на синей кривой  
  
Теперь вернемся к терминам смещение и разброс  
В случайном лесу мы не могли брать рандомные признаки, потому что можно было откидывать какие-то важные признаки и получать смещенную модель, а случайный лес и бэггинг не уменьшают смещение  
А градиентный бустинг может вполне корректировать смещение какой=то одной слабой модели, поэтому мы можем использовать такой подход  
Так даже можно случайно пропустить выброси и снизить ошибку  
  
По графику видно, что оранжевая и синяя кривые это одно и то же, только с добавление семплирования объектов  
Но видно, как семплирование помогает ЕЩЕ снизить ошибку для бустинга с низким весом новой модели
