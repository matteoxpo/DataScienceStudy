Сегодня мы разберем какие модели должны быть в базовой модели, чтобы композиции в бэггинге были сильными  
  
Разберем на уровне идеи без теорвера  
Есть некоторый мат результат, который говорит, что хорошая/плохая модель зависит от 3-х факторов:  
Шум  
Смещение   
Разборс  
  
Рассмотрим пример  
Задача регрессии  
Черная каривая - лучшая модель  
  
Много синих линий - это много разных линейных регрессиий  
Здесь видно, что модель крайне устойчива, но смещение очень большое  
  
Рассмотрим деревья, глубины 2  
Фиолетовая - среднее  
Здесь смещение и разброс как будто поймали дзен  
  
Ну и пример деревьев с разным разборосом и смещением  
Растет глубина - уменьшается смещение, растет разброс  

Можно увидеть, что есть связь между смещением и разбросом композиций и отдельных моделей  
  
Смещение композиции, построенное бэггингом будет такое также же, как и базовых моделей  
С разрбосом все иначе  
Разброс итоговой композиции вычисляется по формуле на слайде  
Это значит, если базовые модели были бы независимы, даже их ошибки независимы, то есть нельзя сказать, ошибется ли другая модель нашей композиции на это объекте, то бэггинг будет сокращать разброс в N раз
Но если все-таки модели связаны, то разброс будет меняться не так активно, за это как раз и отвечает ковариация в формуле.
Итого чем меньше модели похожи друг на друга, тем лучше качество итоговой модели
