каждая из моделей должна быть хотя бы чуть-чуть лучше, чем случайное угадывание - только 1 требование к моделям  
  
В задаче классификации делаем голосованием по большинству  
Это простой способ, выдающих очень хорошее качество  

  
В задаче регрессии берем среднее по нескольким моделям  

А откуда их брать-то?  
  
Второй вариант будет чуть позже, в других лекциях, он называется бустинг:
  
Разберемся с одним из подходов делать деревья независимо и стараемся делать их разнообразными - бэггингом  
  
  
Короче суть простая: ты достал наблюдение в свою выборку и вернул в пул, откуда берешь, то есть можешь взять одно наблюдение несколько раз  

А что если брать выборку не по объектам а по признакам  
  
Обучаем только на этих признаках  
Но если наблюдения относительно равнозначны, а признаки как-то не очень, мы даже не можем так предпологать  
  
Ну и теперь крутим вертим этисм двумя методами в композиции
