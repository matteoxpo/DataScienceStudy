Еще раз уточняется, что для КАЖДОЙ вершины СВОЕ подмножество  
  
  
n_estimators - количество деревьев в лесу  
Как видно по графику после какого-то количества деревьев ошибка заточается в коридор, в плане, перестает меняться  
И причина этому, что разброс уменьшается со временем, а смещение такое же как и у базовых моделей  

Этот метод делает случайный лес самым универсальным методом  
Гиперпараметров толком нет, только размер композиции, но и то, лучше побольше  
Короче просто с коробки все работает в хорошем качестве  
  
Еще несколько особенностей  
Измерение его обучащей способности  
  
$Q_test$ считает прогноз на деревьях, в которые не входил $x_i$ объект, то есть дерево на нем не обучалось  
Очень хорошо можно так оценивать ошибку, практически кросс-валидация  

Оценить насколько важен какой-то признак в предсказании  
  
1. Берем обученный случайный лес
2. Берем тестовую выборку - матрица объекты признаки
3. Берем столбец j-го признака и перемешиваем его среди других признаков, то есть теперь j-ый признак никак не корелирует с другими признаками в наблюдении
4. Мерим качество модели на испорченной выборке
5. Чем хуже стало качество модели, тем важнее признак
