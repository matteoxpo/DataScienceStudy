Вместо 1 категориального заведем m-числовных(бинарных) признаков, первый из этих признаков говорит о том, что в данном наблюдении объект относится к $u_1$ категории, если там 1, то относится, если 0 то не относится<br>
<br>
Так из 1 столбца с M возвожных категорий получится M столбцов с бинарным значением, относится наблюдение к данному признаку или нет<br>


One-hot кодирование называется так, потому что мы выделили M количество бит под кодировку значения, однако всегда максимум активнен 1 бит - горячий<br>
Это значит, что номер бита, который активен соответствует категории<br>
Еще есть One-cold, смысл обратный относительно битов<br>
