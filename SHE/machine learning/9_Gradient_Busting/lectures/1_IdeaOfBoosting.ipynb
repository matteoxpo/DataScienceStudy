Смещение - насколько модель сильна, способна подогнаться под данные  
Разброс - маленький, когда изменения в обучающей выборке не влияют на модель

А разброс уменьшается за счет того, что модели независимы: если 1 модель ошибается в 1 месте, то нельзя ничего сказать про это место у другой модели, то разброс маленький  
  
  
  
# Идея бустинга  
  
Жадно значит, что мы строим каждую следующую модель от текущей так, чтобы она корректировала ошибки уже построенной композиции  
  
$N$ - количество моделей  
$b_n$ - модель под номером n  
Допустим мы построили $N-1$ модель  
Обучение $N$-ой базовой модели  
  
Считаем функцию потерь между уже подсчитанной композицией $a_{N-1}(x_i)$ и новой модели $b_N(x_i)$  
При этом мы строим жадно, что есть мы не вносим корректировок в модели с $1$ по $N-1$ модель, они фиксированы  
Все что мы обучаем это $b_N(x_i)$  
Сложность находится вот тут  
  
Если модель дифференцируема - всё окей, но что делать с деревьями
