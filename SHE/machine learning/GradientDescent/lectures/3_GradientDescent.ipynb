Линия уровня - такая кривая, вдоль которой функция принимает одно и тоже значение, на графике это такие луковицы  

Линия уровня (или изолиния) в контексте градиентного спуска и функций нескольких переменных — это кривая, на которой значение функции остаётся постоянным. То есть, если рассматривать функцию 𝑓(𝑥,𝑦), то линия уровня — это множество точек (𝑥,𝑦) таких, что 𝑓(𝑥,𝑦)=𝑐, где 𝑐 — некоторая константа.
Геометрическая интерпретация: Линия уровня отображает все точки, на которых функция имеет одинаковое значение. В двумерном пространстве (две переменные) линии уровня часто используются для визуализации рельефа поверхности функции. На карте высот, например, линии уровня показывают высоту на уровне моря.

$
\eta_t = \lambda(\frac{s}{s+t})^p
\\
t - номер итерации
\\
s - задает скорость в некотором смысле
\\
\lambda - задает скорость в начале нашего процесса
\\
p - скорость убывания нашего шага
$  
Вернемся к нашей выборке
