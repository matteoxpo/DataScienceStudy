вспоминим ошибки

Посмотрим почему так  
  
Всего 1 выброс так влияет  
  
Мы **испортили модель** на всех нормальных объектах, зато ошибка модели ниже  
Получается выброс важнее нормальных объектов  
Надо как-то решать  
  
  
Пример  
  
Подпортим  
  
То есть со стороны MAE нет смысла портить данные чтобы уменьшить модуль ошибки  
Но модуль не дифференцируется в нуле  
  
Там где модуль отклонения y от a ($|y - a| < \delta$) используем квадрат отклонения  
Идея в том, что функция потерь Хубера ведет себя возле нуля как парабола, а далье как модуль  
  
Ещё одна функция потерь  
  
Иногда надо узнать относительное отклонение прогноза от факта  
$L(y,a)$ - получается как бы процент отклонения в сравнении с целевой переменной  
$Q(a,X) - усредненный$  
  
$y >= 0, a>=0 $
Если $y > a$, то штраф будет не больше 1  
Если $a > y$, то штраф может быть какой угодно    
Это сейчас будет видно на графике  
С точки зрения модели будет получаться, что лучше всего занижать прогноз  
  
График $y = 1$ и варьируем $a$  
Решает эту проблему  
  
  
Больше двух тут штраф не станет
