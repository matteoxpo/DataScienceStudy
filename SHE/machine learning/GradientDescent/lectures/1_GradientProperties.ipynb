# Градиентное обучение  
  
Вспоминим MSE  
  
Градиент это вектор частных производных по всем переменным, от которых зависит функция  
  
  
Выпуклость впуклость функции  
  
Если мы берем любые 2 точки на графике и соединяем их и функция **лежит под прямой**, то такая функция называется **выпуклой**  
А мы всё ещё вспоминаем все  
  
основная проблема что это *очень долго выполняется и не универсально*  
  
Градиент смотри в сторону где функция быстрее всего возрастает  
Анти-градиент(знак минус поставить) показывает обратное, т.е. где функция убывает  
  
  
С помощью антиградиента мы спускаемся к точке минимума - **УРА**  
то есть можно искать где находится минимум какой-то метрики ошибки, например MSE
