Перед Стохастическим спуском вспомним  
  
  
Основная сложность в том, что надо считать градиент для каждого значения на каждом наблюдении  
  
  
А что если вместо среднего по всем градиентом оценим его градиентом 1-го слогаемого по функционалу ошибки  
  
На каждой итерации выбираем случайный объект $i_t$  
Берем значение функции потерь на этом объекте и делаем градиентный шаг  
Дефолтный градиентный спуск   
  
Стохастический  
  
Этого стоило ожидать, траектория туда сюда бегает  
  
И тут мы решаем делать переменную *длину шага*  
  
График ошибки  
  
- Синий - убывание ошибки для полного градиентного спуска  
- Оранжевый - убывание ошибки для стохастического  
Но из плюсов - стохастический очень быстро считается, ведь он всего по 1-му параметру  
  
Модификация, пакетный градиентный спуск  
выбираем $m$-объектов
