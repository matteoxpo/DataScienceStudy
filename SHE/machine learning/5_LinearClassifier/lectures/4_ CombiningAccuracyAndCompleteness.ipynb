## F-мера - Гармоническое среднее  
В общем-то решает поставленный вопрос  
  
По сути есть сглаженная версия минимума  
Рассмаотрим пример  
  
Давайте ее чуть-чуть усложним  
$F_\beta$  
  
По оси Х меняется точность, мы очень быстро меняем линии уровня  
По оси У мы очень долго должны спускаться, чтобы поменять линию уровня  
То есть при $\beta = 0.5$ **важнее точность**  
  
Тут выжнее полнота по понятным причинам
