  
  
Есть еще многоклассовая классификация, но о ней попозже  
  
Очевидное решение  
  
  
  
В случае $<w,x>=0$ это гиперплоскость, которой $w$ перпендикулярен, вообще правильно ортогонален говорить  
  
Другие случаи на слайде расписаны  
  
По сути мы делим нашу выборку некоторой гиперплокостью для классификации  
  
  
Если скалярное произведение близко к нулю, то есть близки к гиперплоскости, то возможно модель сомневается в ответах, поскольку при поялении какого-то шума и смещения гиперплоскости их классификация может измениться  
А также обратим внимание  
  
Для решения этой проблему существует  
  
Если объект положительный и скалярное произведение положительное + * + = +  
Если объект отрицательный и скалярное произведение положительное - * + = -  
Если объект положительный и скалярное произведение отрицательное + * - = -  
То есть если модель правильно угадвает, то отступ положительный
