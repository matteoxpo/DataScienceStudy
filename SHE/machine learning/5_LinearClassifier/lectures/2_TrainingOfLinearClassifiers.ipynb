Рассмаотрим функции потерть для линейного классификатора  
  
Обращу внимание, что функция потерь называется zero-one-loss  
  
У формулы есть куча проблем, обе связаны с недифференцируемостью:  
- Мы берем знак от скалярного произведения, а от этого нельзя взять производную(от знака скалярного произведения по $w$ лол)  
- Берем индикатор, от которого тоже проблемно брать производную  

Решаем эти проблемы  
  
Заменим знак скалярного произведения на условие, что отступ меньше нуля, а бывает только если истинный ответ и прогнозируемы имеют рпзные знаки  
  
По оси Х значение отступа, по оси У значение функции потерь  
Тут функция потерь не дает градиентному спуску понимания куда шагать, допустм мы стоим в точке -1, мини щаг влево вправо ничего не меняет, производная как была равна нулю так и равна нулю, значение самой функции потерь тоже константное  
То есть все-таки надо бы поменять как-то нашу функцию потерь, загнуть ее, чтобы она двигалась как-то иначе  
  
  
То есть основная идея в том, чтобы добавить некоторую верхнюю оценку $\tilde{L}$, функцию, которая находится выше исходной пороговой функции  
По логике эта функция бОльше, чем изначальная и наша задача прижимать максимально вниз новую функцию и надеяться, что так мы понизим ошибку в исходной модели  
  
То есть опять же по сути мы *подменяем* исходную функцию потерь на другую, которая точно не даст ошибку меньше исходной функции  
Пример с использованием логистической верхней оценки  
  
Теперь мы можем делать градиентный спуск  
  
И даже можно регуляризовать все  
  
По сути мы сделали из классификатора с дурацкой функией потерь нормальную такую рабочую модель за счет подмена функции потерь "индикатор ошибки" на верхнюю оценку, которую уже можно нормально дифференцировать
