  
  
Короче посмотрите график сигмоиды чтобы понять че к чему  
  
Хотим на положительных объектах вероятность поближе к 1, на отрицательных поближе к нулю  
  
Обращаю ввнимание на минус перед суммой:  
- Сработал индикатор $y_i = 1$, то есть положительный:  
  Получаем $-\sum_{i=1}^l\sigma(<w, x_i>)$ - то есть чем меньше результат суммы, тем больше значение конечного результата. То есть хтим МАКСИМИЗИРОВАТЬ вероятность положительного объекта  
- Сработал $y_i = -1$, то есть отрицательный:  
  Получаем $\sigma(<w, x_i> - 1)$ если хотим минимизироввать выражение, то нужно минимизировать скалярное произведение, то есть МИНИМИЗИРОВАТЬ вероятность  

 
Но есть проблема такого функционала  
   
Пусть у нас случилась серьзная ошибка  
$y_i = +1$, а модель уверена, что $\sigma(<w, x_i> = 0)$, то штраф равен 1  
Надо в общем еще больше штраф, добавим логарифмы от сигмоиды  
  
$log(0) \rightarrow \infty$ - а вот это уже большой штраф  
Это называется *log-loss*  
Это почти логистическая регрессия, проведем ряд формул  
Исходный функционал с логарифмами  
$-\sum_{i=1}^l[y_i=-1]log\sigma(<w,x_i>) + [y_i=1]log(1 - \sigma(<w,x_i>)) =$  
Подставим сигмоиду  
$-\sum_{i=1}^l[y_i=-1]log\frac{1}{1 + exp(-<w,x_i>)} + [y_i=1]log(1 - \frac{1}{1 + exp(-<w,x_i>)}) =$  
Приведем дробь к общему знаменателю в $log(1 - \frac{1}{1 + exp(-<w,x_i>)})$ и проделаем несложные операции с дробями  
$-\sum_{i=1}^l[y_i=-1]log\frac{1}{1 + exp(-<w,x_i>)} + [y_i=1]log(\frac{1}{1 + exp(<w,x_i>)}) =$  
Воспользуемся свойством логарифма $log(1/x) = -log(x)$ - пропадет минус перед суммой  
$\sum_{i=1}^l[y_i=-1]log{1 + exp(-<w,x_i>)} + [y_i=1]log({1 + exp(<w,x_i>)}) =$  
Слогаемые отличаются только знаком, заметим, что знак перед скалярным произведеним противоположени со знаком индикатора, то есть мы можем просто убрать разделение, заменив знак обратным $y_i$  
$\sum_{i=1}^l log(1 + exp(-y_i \times<w,x_i>))$
