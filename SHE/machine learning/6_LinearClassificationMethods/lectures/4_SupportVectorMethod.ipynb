Хочу обратить внимание, что сначала:  
$w$ и $w_0$ мы делим на $min_{i=1,...,l}|<w,x_i> + w_0|$  
И только потом будет выполняться $min_{i=1,...,l}|<w,x_i> + w_0| = 1$, так как по сути в одном из делений мы получили деление на самого себя  
Также еще мы договаривались, что все объекты можно *ЛИНЕЙНО РАЗДЕЛИМА* и деления на ноль не будет  
Вернемся к подсчету отступа классификатора  
  
Тадам ура наша нормировка помогла найти минимальный отступ  
  
Мы хотим *максимизировать* отступ $\frac{1}{\|w\|} \rightarrow max$  
При условии, что мы выполнили деление всех $w$ и $w_0$ на $min_{i=1,...,l}|<w,x_i> + w_0|$ - мы сделали масштабирование  
Заменим это условие:
Заменим *максимизирование* $\frac{1}{\|w\|}$ на *минимизацию* $\|w\| \rightarrow min$  
и добавим ограничение на то, что $|<w,x_i> + w_0| \geq 1$  
Итого мы можем объеденить требования  
  
Норму вовзели в квадрат для удобства в будущем  
А что делать, если мы не можем нормально разделить все  
  

Мы разрешим иметь на некоторых местах иметь отступ меньше  
  
Но важно подбирать $\xi$ адекватно, чтобы не получилось вот такого  
  
Добавим еще одно ограничение  
  
Коэффициент C должен решать, что важнее:  
C - большое, значит сильно штрафуем за большие значения $\xi$  
C - маленькое, значит важнее минимизировать норму $\|w\|$  
  
  
  
C- гиперпараметр  
Можно сгруппировать условия на $\xi$:  
$y_i(<w, x_i> + w_0) \geq 1 - \xi_i \rightarrow \xi_i \geq 1 -y_i(<w, x_i> + w_0)$  
$\xi_i \geq 0$  
По сути $\xi_i$ должно быть больше либо нуля либо $1 -y_i(<w, x_i> + w_0)$, то есть больше чего-то максимального из этих двух условий  
$\xi_i \geq max(0, y_i(<w, x_i> + w_0))$  
  
  
Со стороный log-loss всегда выгодно сделать отступ всегда побольше, штрафы небольшие, стремит отступы к бесконечности
