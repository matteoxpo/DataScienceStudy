## 1 подход - Калибровка
  
Калибровка - это какая-то новая функция, которая принимает в себя какой-то результат и выпрямляет прямую по вероятностям  
Например у нас есть какой-то выход исходной модели, который плохо оценивает вероятности  
  
Обучим это на логлосс  
  
Наши преобразования  
$c(f) = \frac{1}{1 + exp(p * f * q)}$, где  
$f = b(x)$  
кстати в экспоненте нет минуса, потому что мы надеемся на $p,q$  
и подставим c в логлосс
$-\sum_{i=1}^l [y=+1]log c(f) + [y=-1]log c(1 - f)$  
  
Поэтому используем кросс-валидацию
