Теперь выведим логистическую регрессию иным способом  
Повторим базу  
  
  
Как всегда мы занимаемся обучением модели $b(x)$  
  
Возьмем объекты, у которых вероятности приблизительно равны $p$  
Пусть первые $n$ обхекты это те, у которых примерно одинаковая вероятность, где $n \in p$  
Рассмотрим суммарную ошибку на этих $n$ объектах  
  
Модель выдала вероятность p и она линейная и не может выдать произвольное значение, у нее есть куча ограничений  
Что оптимально выдавать на этих n объектов?  
Необходимо проминизировать функцию суммы потерь  
Мы ожидаем, что $p_*$ - доля положительных объектов среди n объектов, тогда вероятность работает  
Проверим это на log-loss  
  
Найдем минимум  
  
$n_+$ - кол-во объектов с индикатором $[y_i=+1]$  
$n_-$ - кол-во объектов с индикатором $[y_i=-1]$  
  
Мы хотим, чтобы функция потерь требовала от модели, чтобы на объектах с близкими ответами она выдавала вероятность равную примерно доле положительных объектов среди этих n  
<br>
Посмотрим на другие функции ошибки  
  
Требуем чтобы для положительных объектов вероятность была близка к 1 и наоборот  
Но в чем подвох?  
  
Видно, что штраф у log-loss больше, чем больше ошибка  
Штраф у MSE не сильно растет, точнне он около 1 находится  

Возникает проблема с обучением модели, как применить градиентный спуск  

Рассмотрими случай когда всего 1 признак с MSE $b(x) = w_0 + w_1 \times x$  
    
Смотрим график зависимости производной от скалярного произведения  
По графику видно, что производная равна нулю при $w_1 \times x + w_0 \rightarrow \infty$  
И это логично, на прошлом графике видно, что ошибка выходит на асимптоту по штрафу, а значит производная практически равна нулю  
То есть градиентный спуск тут не поможет  

Ну по log-loss все хорошо  
  
По графику все понятно короче, тут градиент уже справится на ура  

Рассмотрим MAE  
  
Окажется, что оптимальное p равно либо нулю, либо единице...  
Так что она вообще тут не подходит
