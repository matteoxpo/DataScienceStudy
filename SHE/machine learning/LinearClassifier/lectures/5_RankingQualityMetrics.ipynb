Как понимать, что модель плохая? или наоборот порог другой надо?  
Посмотрим кривую, которая характеризует $b(x)$  
1 PR-кривая - precision recall  
  
  
то есть каждая точка соответствует значению PR при определенном пороге  
  
Иногда начинают с точки (0;1)  
То есть если идеальная модель проходит через (1;1), то таким образом можно оценить качество модели  
  
Оценим качество как площадь под кривой - AUC-PRC
