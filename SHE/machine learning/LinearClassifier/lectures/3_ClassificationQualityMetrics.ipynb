Проблемы  
  
- Несбалансированные выборки 
Объектов одного класса сильно больше, чем другого  
Например предсказание кликов на экране устройства  
Чаще всего люди не кликают по баннеру, то есть примеров отрицательных будет сильно больше, чем положительных кликов  
И другие примеры на слайде  


  
  
То есть если одного объекта 98 процентов, то доля верных ответов должна быть выше и лежать в диапазоне $accuracy \in [q_o, 1]$, а в моем примере $accuracy \in [0.98, 1]$
